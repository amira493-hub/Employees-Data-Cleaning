# Issues in the data were determined, let's solve them one by one :)

## >>>>> Missing data in the age column

```python
# I didn't just want to fill the missing values with the general mean but has taken the mean of each department region

total_mean= df["Age"].mean()
df["Age"] = df["Age"].fillna( df.groupby("Department_Region")["Age"].transform("mean").astype(int)).fillna(total_mean)
df.head(5)
```


## >>>>> Missing data in the salary column

```python
total_mean= df["Salary"].mean()
df["Salary"] = df["Salary"].fillna( df.groupby("Department_Region")["Salary"].transform("mean").astype(int)).fillna(total_mean)
df.head(5)
```

## >>>>> incorrect symbol in the number column

```python
df["Phone"] = df["Phone"].astype(str).str.replace("-", "")
df.head(3)
```
